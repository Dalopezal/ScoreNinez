Tomamos el modelo de randon forest por que:
es un modelo que se puede utilizar en una amplia gama de problemas de aprendizaje automÃ¡tico, especialmente cuando se necesitan modelos precisos y robustos que puedan manejar conjuntos de datos complejos y de gran escala. aunque en este set de datos no escalamos ni trasnformamos , normalizamos , y  tampoco imputamos datos. ya que fueron generado de forma aleatoria.

Random Forest se basa en Ã¡rboles de decisiÃ³n, que son capaces de manejar datos de forma no lineal y no paramÃ©trica. 

Antes esto es importante aclarar que los datos pueden no seguir una distribuciÃ³n normal para usar Random Forest. se puede  aplicar a conjuntos de datos con cualquier tipo de distribuciÃ³n, ya sean datos categÃ³ricos, ordinales, continuos, etc.

```python
import pandas as pd
df=pd.read_excel('datasets\DimensionSeguridad/PreguntasExperto.xlsx')
```

```python
df.head()
```

```python
# Crear una matriz de correlaciÃ³n para el conjunto de datos "planets"
import seaborn as sns
correlation_matrix = df.corr(numeric_only=True)
# Crear un mapa de calor para visualizar la matriz de correlaciÃ³n
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0)
```

```python
sns.pairplot(data=df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.under_sampling import RandomUnderSampler
import pickle


import pandas as pd
from CargueDatos import *

X = df.drop("Riesgo SG", axis=1)
y = df["Riesgo SG"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

modelo = RandomForestClassifier(n_estimators=100, random_state=42)

modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)

# Evaluar el rendimiento del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

# Mostrar la matriz de confusiÃ³n
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

# Mostrar el informe de clasificaciÃ³n
class_report = classification_report(y_test, y_pred)
print('Classification Report:')
print(class_report)

data = {
    "Edad del menor": 3,
    "Categoria del Municipio": 1,
    "Victima de delito sexual": 1,
    "Victima de violencia intrafamiliar": 2,
    "Victima de conflicto armado": 3,
    "Edad gestora de la madre": 1,
    "Conformacion de su grupo familiar": 2,
}

print(data)
print('prediction')
df_prediction = pd.DataFrame([data])
prediction=modelo.predict(df_prediction)
print(prediction)

### Save model PKL
pck_file = "Seguridad.pkl"
with open(pck_file, 'wb') as file:
    pickle.dump(modelo, file)
```
